## Question 1

<b> Problem Statement - If two people hate each other those two friends will not be invited. maximize the number of people that can attend</b>

## Greedy Solution

<b>Algorithm</b>
<p>The approach idea is to start with the friends who have less enemies so that we can maximize the total number of friends that can attend</p>
<p>so first we sort the sublists based on the length of sublists(from small to big).Then we take the first element and add it to the invite list then we add the elements enemies to hated list so that when the elements in hated list comes up we can skip them. In this way we iterate through all the sublists.</p>

## test cases

[[1,2,3,4],[2,1,5,6,7],[3,1,5,6,7],[4,1,5,6,7],[5,2,3,4,6,7],[6,2,3,4,5,7],[7,2,3,4,5,6]] - Greedy doesnt work<br>
[[0,1,3],[1,0,2],[2,1],[3,4],[4,1]]-both greedy and backtracking work

## CODE

```python
ene_lst1 = [[0,1,3],[1,0,2],[2,1],[3,4],[4,1]]  # in the enemy list first digit signifies the the person and the remaining are his enemies


def greedy(ene_lst1):
    already_hated = set()
    total_participants = len(ene_lst1)
    participants = {i for i in range(total_participants)}
    invite = set()
    #print(ene_lst1)
    ene_lst1.sort(key=lambda x: len(x)) 
    #print(ene_lst1)
    for i,ene in enumerate(ene_lst1):
        for val in ene[1:]:
            if ene[0] not in already_hated:
                temp = ene[0]
            if temp not in already_hated:
                invite.add(temp)
            #print("temp",ene[0])
            for ele in ene_lst1:
                if ele[0]== val:
                    #print('enelst',ele)
                    
                    if temp in ele:
                        if ele[0] not in invite:
                            already_hated.add(ele[0])
        #print('already hated',already_hated)
        #print("invited",invite)
    result = invite
    return result
```

```python
print('The invited people are:',greedy(ene_lst1))
```

## Back Tracking

<b> Algorithm </b>
<p> We can use backtracking to solve the problem </p>
<p> The basic approach is before you add a element you check if there is any other element already present in the invite list that has the current element that is about to be added as an enemy. Then you make two backtracking calls one with tha current element and one without, if the validd function did not satisfy you run a backtracking call with the element. In this way we will be covering the combination from (1,2.....)(1,3.....)..... till (7,2.....).</p>

## CODE

```python
def valid_to_add(person,ene_lst3,invite):
    #print('valid function',person)
    for ele in ene_lst3[person -1][1:]:
        #print('others',ele)
        if ele in invite:
            #print('if in invite',ele)
            return False
    return True
```

```python
def backtracking(ene_lst3,person,invite,total_participants):
  
    if person == total_participants:
        return invite.copy()
    #print('person',person)
    if valid_to_add(person,ene_lst3,invite):
        invite.add(person)
        #print("before",invite)
        result = backtracking(ene_lst3,person+1,invite,total_participants)
        invite.remove(person)

        #print("after",invite)
        result_without = backtracking(ene_lst3,person+1,invite,total_participants)
        if len(result) > len(result_without):
            return result
        else:
            return result_without
    else:
        return backtracking(ene_lst3,person+1,invite,total_participants)
            
    
```

<p> The below given case was the same one which did not give righ answer when greedy approach was used and by using backtracking 
we are able to maximize the people who can be invited</p>

```python
ene_lst3 = [[1,2,3,4],[2,1,5,6,7],[3,1,5,6,7],[4,1,5,6,7],[5,2,3,4,6,7],[6,2,3,4,5,7],[7,2,3,4,5,6]]
invite = set()
total_participants = len(ene_lst3) 
finals = backtracking(ene_lst3,1,invite,total_participants)
```
